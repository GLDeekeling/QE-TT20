## (d)

### (d)i

```r
male_mask = tsa$Gender=="M"
male_mu = mean(tsa$ï..TSA[male_mask])
female_mu = mean(tsa$ï..TSA[!male_mask])
male_sigma = sd(tsa$ï..TSA[male_mask])
female_sigma = sd(tsa$ï..TSA[!male_mask])
male_densities = dnorm(x = xvals,mean = male_mu,sd = male_sigma )
female_densities = dnorm(x = xvals,mean = female_mu,sd = female_sigma )

plot(1,type="n",xlim = c(min(xvals),max(xvals)),ylim = c(0,0.1)) # (d)i
lines(x = xvals,y = female_densities, col="turquoise",lwd = 2)
lines(x = xvals,y = male_densities,col = "peachpuff",lwd = 2)
```

Females have lower scores on average.
